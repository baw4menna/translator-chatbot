using a pre‑trained en to fr model

The translation_en_to_fr pipeline loads anEnglish‑to‑French model such as MarianMT.

```python
translator = pipeline("translation_en_to_fr")
```

2- Building the chatbot

creating a text streaimg function

```python
def typewriter(text: str, delay: float = 0.03):
    for ch in text:
        print(ch, end="", flush=True)
        time.sleep(delay)
    print()
```

The translation pipeline takes english strings and outputs the french translation text

```python
def translate_en_to_fr(text: str) -> str: #Uses the Hugging Face translation pipeline to translate English text into French.
    if not text.strip():
        return ""

    result = translator(text)

    return result[0]["translation_text"]
```

This loop reads english input, calls the translation function, and shows the french result with a typewriter effect.

```python
def run_translation_chat():
    print("English to French Translation Chatbot")
    print("Type 'quit' to exit.\n")

    while True:
        user_input = input("You (EN): ")

        if user_input.lower().strip() == "quit":
            print("Bot (FR): Au revoir !")
            break

        try:
            french = translate_en_to_fr(user_input)
        except Exception as e:
            print(f"Bot: An error occurred: {e}")
            continue

        print("Bot (FR): ", end="")
        typewriter(french, delay=0.03)


run_translation_chat()
```
